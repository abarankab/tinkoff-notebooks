```
import cv2
import face_recognition as face
import matplotlib.pyplot as plt
import matplotlib.patches as patches

%matplotlib inline
```

```
known = face.load_image_file('./my_drive/int/P9010057.jpg')
unknown = face.load_image_file('./my_drive/int/IMG_1515.jpg')
unknown2 = face.load_image_file('./my_drive/int/IMG_1451.jpg')

e1 = face.face_encodings(known)[0]
e2 = face.face_encodings(unknown)[1]
e3 = face.face_encodings(unknown2)[0]
e4 = face.face_encodings(unknown)[0]

r1 = face.compare_faces([e1], e2)
r2 = face.compare_faces([e3], e4)
```

```
print(r1, r2)
```

```
img = unknown2[:]
face_locations = face.face_locations(img)
fig, ax = plt.subplots(1)
ax.imshow(img)

for q in face_locations:
  ax.add_patch(patches.Rectangle((q[3], q[0]), q[1] - q[3], q[2] - q[0], linewidth=1, edgecolor='r', facecolor='none'))

ax.axis('off')
plt.show()
```
